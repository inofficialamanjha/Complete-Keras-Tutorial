# Data Processing

Each data is called a sample. each sample has a label.

In deep learning, sample may be referred as Input data, and Labels may be refered as Target or Output data.

# Sequential Model

Recieves data when we call the 'fit' function on it.

The first two parameter the 'fit' function expects is 'x' and 'y'.

'x' : Input data ( samples ) in the form of 'numpy array', 'tensorflow tensor', a 'dict mapping',a 'tf.data dataset' or a 'keras generator'.

'y' : format - same as x.

Data normalization or standardization can be used to represent data in a understandable form.

Data processing depends on the type of data we are working.

# Data preparation and Processing

```python
# Importing necessary Libraries

import numpy as np
from random import randint
from sklearn.utils import shuffle
from sklearn.preprocessing import MinMaxScaler
```

```python
train_labels = []
train_samples = []
```

# Creating example datset

Example data :

- An experimental drug was tested on individuals from ages 13 to 100 in a clinical trial
- The trial had 2100 participants. Half were under 65 years old, half were 65 years or older.
- Around 95% of patients 65 or older experienced side effects.
- Around 95% of patients under 65 experienced no side-effects

```python
 # the process of creating that dataset
    
for i in range(50) :
    
    # The ~5% of younger individuals who did experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(1)
    
    # The ~5% of older individuals who did not experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(0)
    
for i in range(1000) :
    
    # The ~95% of younger individual who did not experience side effects
    random_younger = randint(13,64)
    train_samples.append(random_younger)
    train_labels.append(0)
    
    # The ~95% of older individual who did experience side effects
    random_older = randint(65,100)
    train_samples.append(random_older)
    train_labels.append(1)
```

```python
for i in range(0,2100):
    print(train_samples[i]," ",train_labels[i])
```

# Data Pre-processing

Making the data fit for model training. Converting it to a numpy array.

```python
train_labels = np.array(train_labels)

train_samples = np.array(train_samples)

train_labels, train_samples = shuffle(train_labels, train_samples) # function declared in Scikit-learn library
```

Either normalize or standardize the data so that we get in such a way that the training of the neural network might become quicker or more efficient.

- Normalization of data :
    
    X' = ( X - X_min ) / ( X_max - X_min )


- Standardization of data :

    X' = ( X - u ) / q
    
    where u : mean
          q : standard deviation

```python
scaler = MinMaxScaler(feature_range = (0,1)) # Scikit-learn object

scaled_train_samples = scaler.fit_transform(train_samples.reshape(-1,1)) # rehsape the array - for 'fit_transform()'
```

```python
for i in scaled_train_samples :
    print(i)
```

```python
scaled_train_samples # an 2-D array of size 1000x1
```

```python
scaled_train_samples.shape # shape (1,)
```

# Creating Neural Network from a Sequential network

We are using Keras API integrated withintensorflow.

i.e, we will build a simple tf.Keras Sequential model

```python
# Importing important modules

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential # used while model building
from tensorflow.keras.layers import Activation, Dense # used while model building
from tensorflow.keras.optimizers import Adam # used while model training
from tensorflow.keras.metrics import categorical_crossentropy # used while training
```

```python
# To check and enable Tensorflow GPU - If running a GPU, run this cell

physical_devices = tf.config.experimental.list_physical_devices('GPU')
print('Num GPUs Availabel : ',len(physical_devices))
tf.config.experimental.set_memory_growth(physical_devices[0],True)
```

```python
# building a sequential model - A linear stack of Layers

model = Sequential([
    Dense(units=16, input_shape = (1,),activation='relu'),
    Dense(units=32,activation='relu'),
    Dense(units=2,activation='softmax') # Fully connected Layer
]) # model is an instance of the 'Sequential' class

# units are also known as nodes or neurons

# softmax gives pro-babilities for each output class
```

```python
model.summary() # Visual representation of our architecture
```

# Training the network

```python
# 'model.compile()' prepares the model for training

# Specify : Optimizer : Learning rate, loss, metrics we want to see ( accuracy, precision, recall)

model.compile(optimizer=Adam(learning_rate=0.0001),loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x=scaled_train_samples, y=train_labels, batch_size=10, epochs=30, shuffle=True, verbose=2) # model converts the labels into one-hot format itself

# verbose level specify the level of output messages we want to see
```

# Visualization and Accuracy Metrics

Used during Inference.

```python
%matplotlib inline

from sklearn.metrics import confusion_matrix
import itertools
import matplotlib.pyplot as plt
```

```python
cm = confusion_matrix(y_true=test_labels,y_pred=rounded_predictions)
```

```python
def plot_confusion_matrix(cm,classes,normalize=False,title='Confusion matrix',cmap = plt.cm.Blues):
    
    
    """
    
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize=True'
    
    """
    
    plt.imshow(cm, interpolation='nearest',cmap=cmap) # defining blue color-maps
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks,classes,rotation=45)
    plt.yticks(tick_marks,classes)
    
    if normalize :
        cm = cm.astype('float') / cm.sum(axis=1)[:,np.newaxis]
        print("Normalized Confusion matrix")
    else:
        print('Confusion matrix, without normalization')
        
    print(cm)
    
    thresh = cm.max()/ 2.
    
    for i,j in itertools.product(range(cm.shape[0]), range(cm.shape[1])) :
        plt.text(j,i,cm[i,j],
                horizontalalignment="center",
                color="white" if cm[i,j] > thresh else "black")
        
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
cm_plot_labels = ['no_side_effects','had_side_effects']
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title = "Confusion matrix")
```
