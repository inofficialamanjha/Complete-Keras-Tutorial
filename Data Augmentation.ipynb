# Data Augmentation

Data-Augmentation occurs when we make a new data using data-modifications to some existing data.

For Image data, data augmentation will include

- Flipping the Image vertically or Horizontally
- Rotating the Image
- Changing the Color of the Image

Adding augmented data can reduce over-fitting as well.

```python
import matplotlib.pyplot as plt
import numpy as np
import os
import random
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator

%matplotlib inline
```

```python
def plotImages(images_arr) :
    fig, axes = plt.subplots(1,10,figsize=(20,20))
    axes = axes.flatten()
    
    for img, ax in zip(images_arr,axes):
        ax.imshow(img)
        ax.axis('off')
        
    plt.tight_layout()
    plt.show()
```

```python
# rotation-range is measured in radians, width-shift-range is measured in percentage of image-width

gen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1,
                        height_shift_range=0.1, shear_range=0.15, zoom_range=0.1,
                        channel_shift_range=10., horizontal_flip=True)
```

```python
chosen_image = random.choice(os.listdir(r'C:\Local Disk\Programming\Python Projects\keras complete\data\MobileNet-samples'))

# chooses and a random image from the list
```

```python
image_path = os.path.join(r"C:\Local Disk\Programming\Python Projects\keras complete\data\MobileNet-samples",chosen_image)
```

```python
assert os.path.isfile(image_path)
```

```python
# Obtain image

image = np.expand_dims(plt.imread(image_path),0) # expands the dimensions of the image
plt.imshow(image[0])
```

```python
# Generate batches of augmented images from this image

aug_iter = gen.flow(image)
```

```python
# Get 10 samples of augmented images

aug_images = [next(aug_iter)[0].astype(np.uint8) for i in range(10)]
```

```python
plotImages(aug_images) # viewing the augmented images
```
